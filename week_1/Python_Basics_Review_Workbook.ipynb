## 🔹 Lists - Practice

**Q: Create a list of numbers from 1 to 5 and print the last element.**

```python
nums = [1, 2, 3, 4, 5]
print(nums[-1])
```

**Q: Remove 'banana' from the list ['apple','banana','cherry'].**

```python
fruits = ['apple','banana','cherry']
fruits.remove('banana')
print(fruits)
```

**Q: Write a list comprehension to create squares [1,4,9,16,25].**

```python
squares = [x**2 for x in range(1,6)]
print(squares)
```

## 🔹 Sets - Practice

**Q: Create a set {1,2,3,3,2,1}. What is the output?**

```python
nums = {1,2,3,3,2,1}
print(nums)
```

**Q: Find intersection of {1,2,3,4} and {3,4,5,6}.**

```python
a = {1,2,3,4}
b = {3,4,5,6}
print(a.intersection(b))
```

**Q: Remove element 4 from {1,2,3} safely.**

```python
nums = {1,2,3}
nums.discard(4)
print(nums)
```

## 🔹 Dictionaries - Practice

**Q: Create dictionary for book with title 'Python 101', author 'John', year 2025.**

```python
book = {'title':'Python 101','author':'John','year':2025}
print(book)
```

**Q: Update year to 2026.**

```python
book['year'] = 2026
print(book)
```

**Q: Get dictionary keys and values separately.**

```python
print(book.keys())
print(book.values())
```

**Q: Swap keys and values using dictionary comprehension for {'a':1,'b':2}.**

```python
d = {'a':1,'b':2}
swapped = {v:k for k,v in d.items()}
print(swapped)
```

## 🔹 Control Flow Tools - Practice

**Q: Write if-elif-else: x=5, print Positive if >0, Zero if =0, Negative if <0.**

```python
x = 5
if x>0:
    print('Positive')
elif x==0:
    print('Zero')
else:
    print('Negative')
```

**Q: Write a for loop to print even numbers from 1 to 10.**

```python
for i in range(1,11):
    if i%2==0:
        print(i)
```

**Q: Write a while loop to print numbers 1 to 3.**

```python
n=1
while n<=3:
    print(n)
    n+=1
```

**Q: Demonstrate break in loop: stop at 5.**

```python
for i in range(1,10):
    if i==5:
        break
    print(i)
```

**Q: Demonstrate continue in loop: skip 3.**

```python
for i in range(1,6):
    if i==3:
        continue
    print(i)
```

## 🔹 File Operations - Practice

**Q: Write 'Hello World' into sample.txt.**

```python
with open('sample.txt','w') as f:
    f.write('Hello World')
```

**Q: Read the contents of sample.txt.**

```python
with open('sample.txt','r') as f:
    print(f.read())
```

**Q: Append 'New Line' to sample.txt.**

```python
with open('sample.txt','a') as f:
    f.write('\nNew Line')
```
