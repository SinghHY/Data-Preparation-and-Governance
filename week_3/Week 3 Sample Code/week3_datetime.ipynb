## 1) Basics: `today()`, `now()`, naive vs aware datetimes

```python
from datetime import datetime, timezone, date, time, timedelta
naive_now = datetime.now()
aware_now_utc = datetime.now(timezone.utc)
print(f"Naive datetime: {naive_now}")
print(f"Aware UTC datetime: {aware_now_utc}")
```

## 2) Constructing `date`, `time`, `datetime`

```python
d = date(2025, 9, 21)
t = time(14, 30, 15, 123456)
dt = datetime.combine(d, t)
d_iso, t_iso, dt_iso = d.isoformat(), t.isoformat(), dt.isoformat()
(d_iso, t_iso, dt_iso)
```

## 4) Arithmetic: `timedelta` addition, subtraction, comparisons

```python

start = date(2025, 1, 15)
duration = timedelta(days=90)
end_date = start + duration
print(end_date)
diff_days = (date(2025, 7, 1) - date(2025, 6, 15)).days

diff_days
```

## 5) Normalizing: zeroing time, start/end of day using `replace`

```python
dt = datetime(2025, 9, 21, 16, 45, 33, 987000)
start_of_day = dt.replace(hour=0, minute=0, second=0, microsecond=0)
end_of_day = dt.replace(hour=23, minute=59, second=59, microsecond=999_999)
(start_of_day, end_of_day)
```

## 11) Business logic: next business day (Monâ€“Fri)

```python

def next_business_day(d: date) -> date:
    d = d + timedelta(days=1)
    while d.weekday() >= 5:  # 5=Sat, 6=Sun
        d += timedelta(days=1)
    return d

ex1 = next_business_day(date(2025, 9, 19))
ex2 = next_business_day(date(2025, 9, 20))
ex3 = next_business_day(date(2025, 9, 21))
(ex1, ex2, ex3)
```
